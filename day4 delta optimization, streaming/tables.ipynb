Clone: Copy of your object (table )

1. Deep Clone 
2. Shallow clone

```python
%sql
CREATE OR REPLACE TABLE dev.chakri.emp_clone_deep
DEEP CLONE dev.chakri.employee;
```

```python
%sql
CREATE OR REPLACE TABLE dev.chakri.emp_clone_shallow
DEEP CLONE dev.chakri.employee;
```

```python
%sql
select * from dev.chakri.emp_clone_shallow
```

```python
%sql
DELETE FROM dev.chakri.emp_clone_shallow WHERE age > 40
```
