```python
schema="Id int ,Name string ,Gender string ,Salary int, Country string ,Date string"
```

```python
(
    spark
    .readStream
    .schema(schema)
    .csv("/Volumes/dev/naval/project/incremental/",header=True)
    .writeStream
    .option("checkpointLocation","/Volumes/dev/naval/project/checkpoint/naval")
    .trigger(availableNow=True)
    .toTable("dev.naval.stream")
    )
```

```python
(
    spark
    .readStream
    .format("cloudFiles") ## autloader 
    .option("cloudFiles.format","csv")
    .option("cloudFiles.schemaLocation", "/Volumes/dev/naval/project/metada/schemalocation/")
    .load("/Volumes/dev/naval/project/incremental/",header=True)
    .writeStream
    .option("checkpointLocation","/Volumes/dev/naval/project/metada/checkpoint/")
    .trigger(availableNow=True)
    .option("mergeSchema",True)
    .toTable("dev.naval.auto")
    )
```
