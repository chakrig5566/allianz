```python
%sql
-- Step 1: Drop the existing column mask on source table so we can read raw emails
ALTER TABLE dev.naval.sales_customers ALTER COLUMN email_address DROP MASK;

-- Step 2: Recreate masked table - only mask emails where continent = 'Asia'
CREATE OR REPLACE TABLE dev.naval.sales_customers_masked AS
SELECT
  customerID,
  first_name,
  last_name,
  CASE
    WHEN continent = 'Asia'
    THEN CONCAT(REPEAT('*', INSTR(email_address, '@') - 1), SUBSTRING(email_address, INSTR(email_address, '@')))
    ELSE email_address
  END AS email_address,
  phone_number,
  address,
  city,
  state,
  country,
  continent,
  postal_zip_code,
  gender
FROM dev.naval.sales_customers;
```

```python
%sql
SELECT customerID, first_name, email_address, continent
FROM dev.naval.sales_customers_masked
WHERE continent IN ('Asia', 'North America')
ORDER BY continent, customerID
LIMIT 50;
```
