##Schema evolution

```python
%python
data=[(1,'Virat'),(2,'Rohit')]
schema="id int, name string"
df=spark.createDataFrame(data,schema)
df.createOrReplaceTempView("source")
```

```python
%sql
select * from source
```

```python
%sql
create table if not exists dev.naval.cricket (id int, name string)
```

```python
%python
df.write.mode("overwrite").saveAsTable("dev.naval.cricket")
```

```python
%sql
insert into dev.naval.cricket select * from source
```

```python
%sql
select * from dev.naval.cricket 
```
